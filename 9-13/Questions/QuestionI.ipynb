```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ProjetsComplet.csv',encoding='utf8',index_col=0)
```

```python
df.sort_values(by = 'Année', axis=0, ascending=True, inplace=True, na_position='last')
df['MontantAccordé'] = pd.to_numeric(df['MontantAccordé'],errors="coerce")
df['MontantAccordéCumulé'] = df.groupby(['Projet'])['MontantAccordé'].cumsum()
df
```

```python
df.sort_values(by = 'Année', axis=0, ascending=True, inplace=True, na_position='last')
```

```python
PHRC
```

```python
PHRC=df[df.Projet == 'PHRC']
PHRIP=df[df.Projet == 'PHRIP']
PREPS=df[df.Projet == 'PREPS']
PRME=df[df.Projet == 'PRME']
PRT=df[df.Projet == 'PRT']

PHRC = PHRC[['Année','MontantAccordéCumulé']]
PHRIP = PHRIP[['Année','MontantAccordéCumulé']]
PREPS = PREPS[['Année','MontantAccordéCumulé']]
PRME = PRME[['Année','MontantAccordéCumulé']]
PRT = PRT[['Année','MontantAccordéCumulé']]

PHRC = PHRC.drop_duplicates(subset='Année', keep="last")
PHRIP = PHRIP.drop_duplicates(subset='Année', keep="last")
PREPS = PREPS.drop_duplicates(subset='Année', keep="last")
PRME = PRME.drop_duplicates(subset='Année', keep="last")
PRT = PRT.drop_duplicates(subset='Année', keep="last")

PHRC['MontantAccordéCumulé'] = PHRC['MontantAccordéCumulé'].div(1000000)
PHRIP['MontantAccordéCumulé'] = PHRIP['MontantAccordéCumulé'].div(1000000)
PREPS['MontantAccordéCumulé'] = PREPS['MontantAccordéCumulé'].div(1000000)
PRME['MontantAccordéCumulé'] = PRME['MontantAccordéCumulé'].div(1000000)
PRT['MontantAccordéCumulé'] = PRT['MontantAccordéCumulé'].div(1000000)

plt.plot(PHRC["Année"], PHRC["MontantAccordéCumulé"],label = 'PHRC')
plt.plot(PHRIP["Année"], PHRIP["MontantAccordéCumulé"],label = 'PHRIP')
plt.plot(PREPS["Année"], PREPS["MontantAccordéCumulé"],label = 'PREPS')
plt.plot(PRME["Année"], PRME["MontantAccordéCumulé"],label = 'PRME')
plt.plot(PRT["Année"], PRT["MontantAccordéCumulé"],label = 'PRT')
plt.legend()
plt.xlabel("Année")
plt.ylabel("Montant cumulé (en millions)")
```
