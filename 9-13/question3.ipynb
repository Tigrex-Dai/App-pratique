```python
import pandas as pd
```

```python
#importation des bases de données
df=pd.read_csv('/content/ProjetsComplet.csv')
df.columns
```

```python
df.columns = ["Unnamed: 0","Finess","Nomdeletablissement","NomInvestigateur","PrenomInvestigateur","TitreProjet","NumeroProjet","Acronyme","MontantAccorde","Annee","Type",
"Projet"]
```

```python
df=df.loc[:,['Annee', 'MontantAccorde','Nomdeletablissement','Projet']]
df
```

```python
df2018=df[df.Annee == 2018]
df2018
```

```python
df2018[pd.to_numeric(df2018.MontantAccorde, errors='coerce').isnull()]
```

```python
df2018['MontantAccorde'] = df2018['MontantAccorde'].astype(float, errors = 'raise')
```

```python
df2018.dtypes
```

```python
df2018['MontantAccorde'] = pd.to_numeric(df2018['MontantAccorde'])
```

```python
df2018=df2018.groupby(['Nomdeletablissement','Annee']).sum()['MontantAccorde']
df2018
```

```python
df2018= df2018.reset_index()
```

```python
df2018=df2018.sort_values('MontantAccorde', ascending = False).head(10)
df2018
```
