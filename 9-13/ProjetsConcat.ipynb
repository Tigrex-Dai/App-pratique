```python
import pandas as pd
import requests
import io
```

```python
urlPHRC = "https://raw.githubusercontent.com/Tigrex-Dai/App-pratique/master/9-13/PHRC/PHRC.csv"
requêtePHRC = requests.get(urlPHRC).content
PHRC = pd.read_csv(io.StringIO(requêtePHRC.decode('utf-8')),index_col=0)
PHRC
```

```python
urlPHRIP = "https://raw.githubusercontent.com/Tigrex-Dai/App-pratique/master/9-13/PHRIP/PHRIP.csv"
requêtePHRIP = requests.get(urlPHRIP).content
PHRIP = pd.read_csv(io.StringIO(requêtePHRIP.decode('utf-8')),index_col=0)
PHRIP
```

```python
urlPREPS = "https://raw.githubusercontent.com/Tigrex-Dai/App-pratique/master/9-13/PREPS/PREPS.csv"
requêtePREPS = requests.get(urlPREPS).content
PREPS = pd.read_csv(io.StringIO(requêtePREPS.decode('utf-8')),index_col=0)
PREPS
```

```python
urlPRME = "https://raw.githubusercontent.com/Tigrex-Dai/App-pratique/master/9-13/PRME/PRME.csv"
requêtePRME = requests.get(urlPRME).content
PRME = pd.read_csv(io.StringIO(requêtePRME.decode('utf-8')),index_col=0)
PRME
```

```python
Projets = pd.concat([PHRC,PHRIP,PREPS,PRME])
Projets
```
